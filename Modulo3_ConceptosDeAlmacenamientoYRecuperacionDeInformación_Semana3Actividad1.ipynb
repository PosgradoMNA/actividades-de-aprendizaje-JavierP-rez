### Fundamentos de bases de datos y para ciencia de datos.

Una base de datos es una colección organizada de datos, lógicamente relacionados. Para el apartado de Ciencia y analítica de datos se hablará de bases de datos relacionales SQL. La información se organiza por tablas, es un conjunto de columnas y filas (o campos o registros); a la intersección entre columnas y filas se le conoce como celda. Un gestor de base de datos es un software que permite administrar y mantener la base de datos; por ejemplo, consultar, crear, accesos mediante un lenguaje estructurado llamado SQL. Tener un base de datos es una ventaja de negocio, debido a que permite organizar y consultar la información de una manera más rápida y eficiente; de igual, implementar técnicas o procesos como AI, Big Data, etc. En mi experiencia, he trabajado con bases de datos Oracle.

### Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Data Warehouse es una base de datos grande que contiene información de diferentes fuentes y las reúne; de esta manera, se procesa la información para usarla con otras tecnologías. De ahí existen procesos para cargar la información a estos sistemas, como ETL o ELT (sus siglas significan extraer, transformar y cargar o extraer, cargar y transformar).  Mediante frameworks, lenguajes de programación, se pueden llevar procesos de extracción, procesamiento y carga de información. Hoy en día, existen varios sistemas en dónde se pueden almacenar estos sistemas; el más recurrente por las empresas, que está aumentando su popularidad, es el de Nube. Muchas empresas, como Amazon, Microsoft, Google, Oracle o más, desarrollaron su propia infraestructura en nube y ofrecen servicios de Warehouse, para el mejor procesamiento de datos y obtención de estadísticas y así  mejorar los procesos del producto/servicio.

### Solución 2

Se crea el data frame con los datos. Y se obtiene el promedio, junto con algunos valores de la tabla. 

```python
import pandas as pd
import numpy as np

csv_path = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
df = pd.read_csv(csv_path, index_col=0)

ndf = df.copy()
```

```python
ndf
```

```python
wm = ndf.X1
wm
```

```python
wm = ndf.X1.mean()
wm
```

```python
ndf['X2']
```

Esta línea de código, se llenan los valores de la columna X1 con NA/NaN con el valor promedio. Se modifica la tabla original. 

```python
ndf['X1'].fillna(value = wm, 
                    inplace = True)
ndf
```

Se llenan los valores con la mediana, pero de la columna X2. 

```python
ndf['X2'].fillna(value = ndf.X2.median(), 
                    inplace = True)
ndf
```

Se indica el valor que más aparece en la columna, junto con su tipo. 

```python
mm  = ndf.X1.mode()
mm
```

```python
mm[0]
```

Se rellena la tabla con el valor más repetido de la columna. 

```python
ndf['X1'].fillna(value = mm[0], #'NoPais', 
                    inplace = True)
ndf
```

Se descubre si hay valores vacíos. 

```python
ndf.isnull().values.any()
```

```python
ndf = df.copy()
ndf
```

Se eliminan las filas que contengan valores Nan en las columnas X1 y X2.

```python
ndf.dropna(subset=['X1', 'X2'], inplace = True)
ndf
```

```python
ndf = df.copy()
ndf
```

Se obtiene el valor más repetido. 

```python
ndf.X1.mode()[0]
```

Se crea un diccionario, con las llaves que son el nombre de las columnas. Y de valores el valor más repetido, y el promedio. 

```python
favs = {'X1': ndf.X1.mode()[0], 'X2': ndf['X1'].mean()}
print(favs)
```

Se llenan los valores faltantes con el valor más repetido y el promedio, en las filas X1 y X2.

```python
ndf.X1.fillna(ndf.X1.mode()[0], inplace=True)
ndf.X2.fillna(ndf.X2.mean(), inplace=True)
ndf
```
