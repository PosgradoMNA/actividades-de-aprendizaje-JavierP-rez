```python
from ast import increment_lineno
import numpy as np

#List. 
l = [1, 2, "a"]

#Array
a = np.array([0, 1, 2, 3, 4, 5])
print(a,type(a))
print(a.size)
print(a.ndim)
print(a.shape)

b = np.array([10.3, 1.3, 2.5, 3.8, 4.4, 5.2])
print(b, type(b))
print(b.dtype)

#Indexing and Slicing.
c = np.array([0, 1, 2, 3, 4, 5])
c[0] = 100
print(c)
c[4] = 0
print(c)
print(c[1:4])
c[3:5] = 300,400
print(c)

#Operations

u = np.array([2,1])
v = np.array([3,4])

print(u + v)

u = [1,0]
v = [1,0] 

z = []

for n,m in zip(u,v):
  z.append(n+m)

u = np.array([1,0])
v = np.array([0,1])

z = u + v

print(z, type(z))

z = u - v

print(z, type(z))

v = [1,0] 
print(2*v)


u = np.array([15,3])
v = np.array([2,4])

print(u*v)

result = np.dot(u,v)
print(result)

#Universal Functions
a = np.array([1, -1, 1, -1, 2, 1, 6, 7, 9, 5])
mean_a = a.mean()
print(mean_a)
print(b.max())

print(np.pi)
print(np.sin(90))

print(np.linspace(-2,2,num=5))

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x,y)
```

```python
#Convert the list [1, 2, 3, 4, 5] and [6, 7, 8, 9, 10] to numpy arrays arr1 and arr2. Then find the even and odd numbers from arr1 and arr2.

arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])

x = arr1[1:8:2]
x
   
y = arr2[0:8:2]
y   

```

```python
#Convert the list [1, 2, 3] and [8, 9, 10] to numpy arrays arr1 and arr2. Then perform Addition , Subtraction , Multiplication , Division and Dot Operation on the arr1 and arr2.
arr1 = np.array([1, 2, 3])
arr2 = np.array([8, 9, 10])

arr3 = np.add(arr1, arr2)
arr3

arr4 = np.subtract(arr1, arr2)
arr4

arr5 = np.multiply(arr1, arr2)
arr5


arr6 = np.divide(arr1, arr2)
arr6

arr7 = np.dot(arr1, arr2)
arr7
```
